```python
# --- Clean Customer Table ---
customer_df = df[[
    "Customer.ID", "Customer.Name", "Segment", "Country", "City", "State", "Region", "Market"
]].copy()

# Normalize Customer IDs
customer_df["Customer.ID"] = (
    customer_df["Customer.ID"]
    .astype(str)          # ensure string type
    .str.strip()          # remove leading/trailing spaces
    .str.upper()          # standardize case
)

# Drop duplicates based on Customer.ID
customer_df = customer_df.drop_duplicates(subset=["Customer.ID"])

# Save cleaned file
customer_df.to_csv("customers.csv", index=False)

# --- Clean Product Table ---
product_df = df[[
    "Product.ID", "Product.Name", "Category", "Sub.Category"
]].copy()

product_df["Product.ID"] = (
    product_df["Product.ID"]
    .astype(str)
    .str.strip()
    .str.upper()
)

product_df = product_df.drop_duplicates(subset=["Product.ID"])
product_df.to_csv("products.csv", index=False)

# --- Clean Sales Table ---
sales_df = df[[
    "Row.ID", "Order.ID", "Order.Date", "Ship.Date", "Ship.Mode",
    "Customer.ID", "Product.ID", "Order.Priority", "Year", "weeknum",
    "Sales", "Quantity", "Discount", "Profit", "Shipping.Cost", "Market2"
]].copy()

# Normalize foreign keys
sales_df["Customer.ID"] = sales_df["Customer.ID"].astype(str).str.strip().str.upper()
sales_df["Product.ID"] = sales_df["Product.ID"].astype(str).str.strip().str.upper()

sales_df.to_csv("sales.csv", index=False)
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS Customers (
    CustomerID VARCHAR(50) PRIMARY KEY,
    CustomerName VARCHAR(255),
    Segment VARCHAR(100),
    Country VARCHAR(100),
    City VARCHAR(100),
    State VARCHAR(100),
    Region VARCHAR(100),
    Market VARCHAR(100)
);
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS Products (
    ProductID VARCHAR(50) PRIMARY KEY,
    ProductName VARCHAR(255),
    Category VARCHAR(100),
    SubCategory VARCHAR(100)
);
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS Sales (
    RowID INT PRIMARY KEY,
    OrderID VARCHAR(50),
    OrderDate DATE,
    ShipDate DATE,
    ShipMode VARCHAR(100),
    CustomerID VARCHAR(50),
    ProductID VARCHAR(50),
    OrderPriority VARCHAR(50),
    Year INT,
    WeekNum INT,
    Sales DECIMAL(18,2),
    Quantity INT,
    Discount DECIMAL(5,2),
    Profit DECIMAL(18,2),
    ShippingCost DECIMAL(18,2),
    Market2 VARCHAR(100),
    FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID),
    FOREIGN KEY (ProductID) REFERENCES Products(ProductID)
);
""")
```

```python
import pandas as pd

# Load cleaned CSV
customers = pd.read_csv("customers.csv")

# Prepare insert query
insert_customer = """
INSERT INTO Customers (CustomerID, CustomerName, Segment, Country, City, State, Region, Market)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
"""

# Convert DataFrame to list of tuples
customer_data = [tuple(x) for x in customers.to_numpy()]

# Bulk insert
cursor.executemany(insert_customer, customer_data)
conn.commit()
```

```python
products = pd.read_csv("products.csv")

insert_product = """
INSERT INTO Products (ProductID, ProductName, Category, SubCategory)
VALUES (%s, %s, %s, %s)
"""

product_data = [tuple(x) for x in products.to_numpy()]
cursor.executemany(insert_product, product_data)
conn.commit()
```

```python
sales = pd.read_csv("sales.csv")

insert_sales = """
INSERT INTO Sales (RowID, OrderID, OrderDate, ShipDate, ShipMode, CustomerID, ProductID,
                   OrderPriority, Year, WeekNum, Sales, Quantity, Discount, Profit, ShippingCost, Market2)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

sales_data = [tuple(x) for x in sales.to_numpy()]
cursor.executemany(insert_sales, sales_data)
conn.commit()
```
